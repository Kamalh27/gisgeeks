### Import libraries

```python
import netCDF4
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# Define file location
file_path = "data/tasmax_day_GFDL-CM4_ssp245_r1i1p1f1_gr1_2022.nc"
```

```python
# Read this data using NetCDF4
nc = netCDF4.Dataset(file_path, mode="r")
print(nc)
```

```python
# Check the data type
print(type(nc))
```

```python
# Explore the variables
print(nc.variables.keys())
```

```python
# Explore specific variable
print(nc["tasmax"])
```

```python
# Plotting data using matplotlib
plt.contourf(nc['tasmax'][0,:,:])
plt.colorbar()
```

```python
# Change the default colormap
plt.contourf(nc['tasmax'][0,:,:], cmap='jet')
plt.colorbar()
```

```python
# Define figure size
plt.figure(figsize=(16,10))

# Add label and titile of figure
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Map of Maximum Temperature (K), 2022")

plt.contourf(nc['tasmax'][0,:,:], cmap='jet')
plt.colorbar()
```
